```python
import pandas as pd
import re
import string
 

data = pd.read_csv("Data/spambase.csv", sep=',', header=None)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.model_selection import train_test_split
```

```python
email_class = data[57]
```

```python
data.drop([57],axis = 1, inplace=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, email_class, test_size=0.25)
```

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
alg_RandomForest = RandomForestClassifier()
params = {'n_estimators': [5, 20, 50,100],
        'max_depth': [10, 20, 30]}

hyper_params_grid = GridSearchCV(alg_RandomForest, params, cv=4, n_jobs=-1)
hyper_params_models = hyper_params_grid.fit(x_train, y_train)
```

```python
pd.DataFrame(hyper_params_models.cv_results_).sort_values('mean_test_score', ascending=False)[0:6]
```
